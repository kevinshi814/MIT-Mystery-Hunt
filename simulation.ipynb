```python
import numpy as np
import pandas as pd

from sklearn.metrics.pairwise import haversine_distances
```

```python
matches = []
for i in range(1, 7):
    match = pd.read_csv(f"Match{i}.csv")
    matches.append(match)
```

```python
dists = []
for match in range(6):
    match_dists = []
    for ship in range(len(matches[match])):
        match_dists.append(round(180 / np.pi * haversine_distances(
            [[matches[match].at[ship, 'Start Lat'] * np.pi/180,
              matches[match].at[ship, 'Start Lon'] * np.pi/180]],
            [[matches[match].at[ship, 'End Lat'] * np.pi/180,
              matches[match].at[ship, 'End Lon'] * np.pi/180]])[0][0], 3))
    dists.append(match_dists)

dists
```

```python
def distance(lat1, lon1, lat2, lon2):
    return haversine_distances(
        [[lat1 * np.pi/180, lon1 * np.pi/180]], [[lat2 * np.pi/180, lon2 * np.pi/180]]
    )[0][0] * 180 / np.pi
```

```python
dists = []
for match in range(6):
    match_dists = []
    for ship in range(len(matches[match])):
        match_dists.append(round(distance(matches[match].at[ship, 'Start Lat'], matches[match].at[ship, 'Start Lon'],
                                          matches[match].at[ship, 'End Lat'], matches[match].at[ship, 'End Lon']), 3))
    dists.append(match_dists)

dists
```

```python
def decode(n):
    n %= 29
    if n == 0:
        return 'ED'
    elif n <= 26:
        return chr(ord('A') + n - 1)
    elif n == 27:
        return 'BE'
    else:
        return 'LI'
```

```python
def get_position(match, time):
    lat1, lon1 = match['Start Lat'] * np.pi/180, match['Start Lon'] * np.pi/180
    lat2, lon2 = match['End Lat'] * np.pi/180, match['End Lon'] * np.pi/180

    x1 = np.cos(lat1) * np.cos(lon1)
    y1 = np.cos(lat1) * np.sin(lon1)
    z1 = np.sin(lat1)
    x2 = np.cos(lat2) * np.cos(lon2)
    y2 = np.cos(lat2) * np.sin(lon2)
    z2 = np.sin(lat2)

    t_omega = time/90 * np.pi/2
    sin = np.sin(t_omega)
    cos = np.cos(t_omega)

    # simplified slerp
    x = cos*x1 + sin*x2
    y = cos*y1 + sin*y2
    z = cos*z1 + sin*z2

    lat = np.arctan2(z, np.sqrt(x**2 + y**2))
    lon = np.arctan2(y, x)
    return lat * 180/np.pi, lon * 180/np.pi
```

```python
ans = ""
for match in matches:
    num_ships = len(match)

    hits = [[False] * num_ships for _ in range(num_ships)]

    for ship in range(num_ships):
        t = match.at[ship, 'Launch Time']
        radius = match.at[ship, 'Blast Radius']
        ship_locs = get_position(match, t)
        for other in range(num_ships):
            if distance(ship_locs[0][other], ship_locs[1][other],
                        match.at[ship, 'Target Lat'], match.at[ship, 'Target Lon']) <= radius:
                hits[ship][other] = True

    ship_active = [True] * num_ships
    bounty_claimed = [False] * num_ships
    for ship in range(num_ships):
        bounty = 0
        if ship_active[ship]:
            for other in range(num_ships):
                if hits[ship][other] and not bounty_claimed[other]:
                    ship_active[other] = False
                    bounty_claimed[other] = True
                    bounty += match.at[other, 'Bounty']
        if bounty > 0:
            print(decode(bounty))
            ans += decode(bounty)

    print()
    ans += " "
```
