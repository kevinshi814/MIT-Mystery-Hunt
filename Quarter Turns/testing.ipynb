```python
import numpy as np
import pandas as pd
```

```python
match4 = pd.read_csv(f"Match4.csv")
match4.loc[4, 'Start Lon'] += 6.11
match4.loc[7, 'Target Lon'] += 6.11

match4
```

```python
hits = [[False] * 8 for _ in range(8)]

for ship in range(8):
    t = match4.at[ship, 'Launch Time']
    radius = match4.at[ship, 'Blast Radius']
    for other in range(8):
        if abs(match4.at[other, 'Start Lon'] + t - match4.at[ship, 'Target Lon']) <= radius:
            hits[ship][other] = True

hits
```

```python
def decode(n):
    n %= 29
    if n == 0:
        return 'ED'
    elif n <= 26:
        return chr(ord('A') + n - 1)
    elif n == 27:
        return 'BE'
    else:
        return 'LI'
```

```python
ship_active = [True] * 8
bounty_claimed = [False] * 8
for ship in range(8):
    bounty = 0
    if ship_active[ship]:
        for other in range(8):
            if hits[ship][other] and not bounty_claimed[other]:
                ship_active[other] = False
                bounty_claimed[other] = True
                bounty += match4.at[other, 'Bounty']
    if bounty > 0:
        print(decode(bounty))
```
